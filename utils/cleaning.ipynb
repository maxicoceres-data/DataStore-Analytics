### Precios

```python
# Trato de precios, si los precios son menores a 0 es incorrecto, eliminar o solucionarlos.

precios_negativos = df[df["precio_unitario"] <= 0]
filas_precios_negativos = len(precios_negativos)
filas_df = len(df)


if filas_precios_negativos < (filas_df * 0.1):
    df = df[df["precio_unitario"] > 0]
    print(f"Filas descartadas: {filas_precios_negativos}")
else:
    print(f"Demasiados precios inválidos ({filas_precios_negativos} / {filas_df}). Revisar el archivo.")
```

### Corrección datos nulos en cantidad

```python
cantidad_null_negativo = df.loc[(df["cantidad"]<=0) | (df["cantidad"].isna())]
filas_cantidad_null_negativo = len(cantidad_null_negativo)

if filas_cantidad_null_negativo < df.shape[0] * 0.1:
    df = df[df["cantidad"] > 0]
    print(f"Filas eliminadas: {filas_cantidad_null_negativo}, por elemenos nulos o negativos, menos del 10% del dataset.")
else:
    print(f"Cantidad erronea alta. ({df.shape[0] / filas_cantidad_null_negativo}), revisar el dataset.")
```

### Corrección nulos de productos

```python
df_productos_nulos = df.loc[df["producto"].isnull()]
```

```python
### valido los nombres de los productos, y hago una referencia de estos productos agrupandolos por categoria y producto y haciendo la mediana de los mismos para restar.

validos = ["Remera básica", "Jean slim", "Zapatillas running", "Campera impermeable",
           "Vestido floral", "Short deportivo", "Buzo hoodie", "Medias pack x3",
           "Gorra snapback", "Cinturón cuero"]

ref = (
    df[df["producto"].isin(validos)]
    .groupby(["categoria","producto"])["precio_unitario"]
    .median()
)

ref
```

```python
# Hago un bucle for para recorrer las filas de los productos nulos, encontrar el precio de esos productos, restarles el precio con los precios de las medianas de la referencia, y los que son mas bajos (mas cercano a 0) reemplazo el nombre del producto en el nan. 

mask_invaidos = ~df["producto"].isin(validos)

for i in df[mask_invaidos].index:
    categoria = df.loc[i,"categoria"]
    precio = df.loc[i,"precio_unitario"]
    candidatos = ref.loc[categoria]
    df.loc[i,"producto"] = (candidatos - precio).abs().idxmin()
```

```python
df["producto"].unique()
```

```python

#normalizo los nombres de los productos poniendo en minusculas. 
productos_normalizado = {
    'Gorra snapback':'Gorra snapback',
    'Remera básica': 'Remera basica',
    'Vestido floral':"vestido floral",
    'Zapatillas running':'zapatillas running',
    'Cinturón cuero':"cinturon cuero",
    'Buzo hoodie' :"buzo hoodie",
    'JEAN SLIM':'Jean slim',
    'Short deportivo' : "short deportivo",
    'Medias pack x3' : "medias pack x3",
    'Campera impermeable':'campera Impermeable',
}


df["producto"] = df["producto"].replace(productos_normalizado).str.lower()

df["producto"].unique()
```
